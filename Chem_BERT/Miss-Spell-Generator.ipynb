```python
with open(file) as fp:
    data = fp.read().lower()
```

```python
data = data.replace('"',"").splitlines()
len(data)
```

```python
def normalize(each_string):
    each_string = re.sub(r'[^A-Za-z\(\),\[\] ]+', '-', each_string)
    each_string = re.sub(r'\(+|\)+|,+|\[+|\]+|\[-\]', '-', each_string) 
    each_string = re.sub(r'^[-| ]|[-| ]+$', "-", each_string)
    each_string = re.sub(r'- -+|- +| -+|--+', '-', each_string)
    each_string = re.sub(r'- -+|- +| -+|--+', '-', each_string)
    each_string = re.sub(r'^[-]+|[-]+$', "", each_string)
    return each_string.strip()
```

```python
replacevouls = ["a","e","i","o","u","y","ui","ie","uo", "ou"]
replacepunction = ["(",")","[","]","-",",",".",":",";","%","&","*","$","#","@","_","+","=","~","`","<",">","/","?"]
```

```python
datadict = {}
for each in data:
    
    each = each.strip()
    norm = normalize(each)
    lists = [s.strip() for s in re.split("-", norm)]
    norm = max(lists, key = len)
    datadict[each] = norm
    datadict[norm] = norm
    datadict[re.sub(r'[^a-z ]+', '', norm)] = norm
    
        
    for punct in replacepunction:
        datadict[each.replace(punct, "").strip()] = norm
        datadict[each.replace(punct, " ").strip()] = norm
        datadict[each.replace(punct, "-").strip()] = norm
        
    splitnorm = list(norm)
    spliteach = list(each)
    for item in replacevouls:
        if len(splitnorm)>5:
            datadict[each.replace(item, "").strip()] = norm
        if len(splitnorm)>7:
            datadict[each.replace(item, " ").strip()] = norm
            
    if len(splitnorm)>15:
        datadict["".join(spliteach[0:-7]).strip()] = norm
        datadict["".join(spliteach[0:-6]).strip()] = norm
        datadict["".join(splitnorm[0:-7]).strip()] = norm
        datadict["".join(splitnorm[0:-6]).strip()] = norm
        
        datadict["".join(splitnorm[1:-7]).strip()] = norm
        datadict["".join(splitnorm[1:-6]).strip()] = norm
        datadict["".join(spliteach[1:-5]).strip()] = norm
        datadict["".join(spliteach[1:-4]).strip()] = norm
        
        datadict["".join(splitnorm[2:-6]).strip()] = norm
        datadict["".join(splitnorm[2:-5]).strip()] = norm
        datadict["".join(spliteach[2:-4]).strip()] = norm
        datadict["".join(spliteach[2:-3]).strip()] = norm
        
    if len(splitnorm)>12:
        datadict["".join(splitnorm[0:-5]).strip()] = norm
        datadict["".join(splitnorm[0:-4]).strip()] = norm
        datadict["".join(spliteach[0:-5]).strip()] = norm
        datadict["".join(spliteach[0:-4]).strip()] = norm
        
        datadict["".join(splitnorm[1:-5]).strip()] = norm
        datadict["".join(splitnorm[1:-4]).strip()] = norm
        datadict["".join(spliteach[1:-3]).strip()] = norm
        datadict["".join(spliteach[1:-2]).strip()] = norm
        
        datadict["".join(spliteach[2:-2]).strip()] = norm
        datadict["".join(spliteach[2:-1]).strip()] = norm
        
    if len(splitnorm)>10:
        datadict["".join(splitnorm[0:-3]).strip()] = norm
        datadict["".join(splitnorm[0:-2]).strip()] = norm
        datadict["".join(spliteach[0:-2]).strip()] = norm
        datadict["".join(spliteach[1:-1]).strip()] = norm
        
    if len(splitnorm)>5:
        datadict["".join(splitnorm[0:-1]).strip()] = norm
        datadict["".join(splitnorm[1:-1]).strip()] = norm
        datadict["".join(spliteach[0:-2]).strip()] = norm
        datadict["".join(spliteach[1:-1]).strip()] = norm
    
    randomChar = random.choice(splitnorm).strip()
    randomVoul = random.choice(replacevouls).strip()
    if len(spliteach)>20 and randomChar in each and randomVoul in each:
        datadict[each.replace(randomChar,randomVoul+randomChar)] = norm
    
    randomChar = random.choice(spliteach).strip()
    randomVoul = random.choice(replacevouls).strip()
    if len(spliteach)>7 and randomChar in each and randomVoul in each:
        datadict[each.replace(randomVoul,randomChar)] = norm
        datadict[each.replace(randomChar, randomVoul)] = norm
        
    randomChar = random.choice(spliteach).strip()
    if len(spliteach)>5 and randomChar in each:
        datadict[each.replace(randomChar,"").strip()] = norm
        datadict[each.replace(randomChar," ").strip()] = norm
        datadict[each.replace(randomChar,"-").strip()] = norm
        
        
```
